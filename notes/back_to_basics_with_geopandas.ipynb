We're going to look at the EPA Ecoregions for this: https://www.epa.gov/eco-research/level-iii-and-iv-ecoregions-continental-united-states

```python
epa_files = gpd.read_file("C:/Users/thewe/OneDrive/Desktop/College/AES408/files/us_eco_l3_state_boundaries")
```

```python
for index, row in epa_files.iterrows():
    print(f"{index},{row["EPA_REGION"]}")
    break
```

```python
epa_region_counts = dict()
for index, row in epa_files.iterrows():
    cur_epa_reg = row['EPA_REGION']
    if cur_epa_reg in epa_region_counts:
        epa_region_counts[cur_epa_reg]+=1
    else:
        epa_region_counts[cur_epa_reg] = 1

epa_region_counts
```
