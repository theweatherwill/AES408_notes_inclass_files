```python
# Works for vector and raster data
# Not nearly as flexible as what is built in (can't mix int and str)
# numpy is often in the backend of things
import numpy as np
```

```python
# numpy will use Python lists and spits it out as a numpy array.
x = [1,2,3]
test_arr = np.array(x)
```

```python
test_arr
```

```python
# Python can't do this! Scalar operation like a vector. (1 + 3 = 4, 2 + 3 = 5, etc.)
test_arr + 3
```

```python
test_arr ** 2
```

```python
# You can add numpy arrays of the same shape. If you try to add [x,y,z] and [x,y] you'll get an error.
test_arr + [3,4,5]
```

```python
# arange works like (start,stop,step)
test_arr = np.arange(0,100, 2)
test_arr
```

```python
# creates a 20x20 matrix that is just full of zeros for some reason
test_arr = np.zeros((20,20))
test_arr
```

```python
# linspace = return evenly space samples, bounded by square brackets on both sides. Good for color plots!
# this example, provides 5 evenly spaced values between 30 and 40
test_arr_1 = np.arange(0,10)
test_arr_2 = np.linspace(30,40,5)
test_arr_2
```

```python
# concatenate smacks the two arrays side by side and boom they're one array. This case, they're not assigned to a new array
np.concat([test_arr_1, test_arr_2])
```

```python
# axis=0 just puts them together: regular concat
x = np.array([[1,2],[3,4]])
y = np.array([[5,6]])
np.concatenate((x,y),axis=0)
```

```python
# axis=1 looks at the same position of each array and creates a new array from those.
# for this example, [1,2] and [5,6] are both position 0, so it'll combine position 0 values
x = np.array([[1,2],[3,4]])
y = np.array([[5,6],[7,8]])
np.concatenate((x,y),axis=1)
```
