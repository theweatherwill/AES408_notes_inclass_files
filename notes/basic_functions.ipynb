```python
def add_all(*args):
    total=0
    for arg in args:
        total+=arg
    return total
add_all(2,3,4,5)
```

```python
# args give tuples, kwargs give dicts
# must have a keyword argument
    # ex: a=4, b=5....... 4,5 won't work
def add_all(**kwargs):
    total = 0
    for key in kwargs:
        total+=kwargs[key]
    return total
add_all(a=4,b=5)
```

```python
# both valid
print(add(1,2,3))
print(add(3,4,5,6,7,2,1,2,3,4))
```

```python
# from the python documentation

def cheeseshop(kind, *arguments, **keywords):
    print("-- Do you have any", kind, "?")
    print("-- I'm sorry, we're all out of", kind)
    for arg in arguments:
        print(arg)
    print("-" * 40)
    for kw in keywords:
        print(kw, ":", keywords[kw])
```

```python
cheeseshop("Nothing Burger", "Was it crispy?",
           "Right right...",
           shopkeeper="Michael Palin",
           client="John Cleese",
           sketch="Cheese Shop Sketch")
```

```python
# from the Python documentation
def standard_arg(arg):
    print(arg)

def pos_only_arg(arg, /):
    print(arg)

def kwd_only_arg(*, arg):
    print(arg)

def combined_example(pos_only, /, standard, *, kwd_only):
    print(pos_only, standard, kwd_only)
```
